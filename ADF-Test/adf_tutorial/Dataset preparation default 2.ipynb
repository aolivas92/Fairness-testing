```python
import sys
import numpy as np
sys.path.append('../')
import pandas as pd
from sklearn.preprocessing import StandardScaler,MinMaxScaler
import tensorflow.compat.v1 as tf 
tf.compat.v1.disable_eager_execution()

from sklearn.preprocessing import LabelEncoder 
le  = LabelEncoder()
std = MinMaxScaler()
#std = StandardScaler()
df=pd.read_csv('../datasets/default_raw/Default_credit.csv')
df=df.drop(columns=['ID'])
df['SEX'] = le.fit_transform(df['SEX'])
df['EDUCATION'] = le.fit_transform(df['EDUCATION'])
df['MARRIAGE'] = le.fit_transform(df['MARRIAGE'])
bins = np.linspace(df['AGE'].min(), df['AGE'].max(), 10)
df['AGE'] = pd.cut(df['AGE'], bins, labels =[1,2,3,4,5,6,7,8,9], include_lowest=True )
for col in ['PAY_0', 'PAY_2','PAY_3', 'PAY_4', 'PAY_5', 'PAY_6']:
    df[col] = le.fit_transform(df[col])

for col in list(df.columns[11:-1]) + ['LIMIT_BAL']:
    df[col] = (std.fit_transform(df[col].to_numpy().reshape(-1,1)) * 100 ).astype('int')

#     bins = np.linspace(df[col].min(), df[col].max(), 20)
#     df[col] = pd.cut(df[col], bins, labels =list(range(1,20)), include_lowest=True )
# df = df.astype('int')
df.to_csv('../datasets/Default_credit_clean.csv',index=False)
```
